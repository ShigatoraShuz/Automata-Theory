```python
# ============================================================
# CLEAN FINAL PIPELINE
# TOP 3 LINEAR SVM PERFORMERS ONLY
# TF-IDF WORD + CHAR FEATURES
# SEABORN CONFUSION MATRIX
# PER-CATEGORY ACCURACY, RECALL, F1, SUPPORT
# ============================================================


# ============================================================
# 1. INSTALL LIBRARIES
# ============================================================

!pip install -q kagglehub scikit-learn pandas numpy matplotlib seaborn joblib


# ============================================================
# 2. IMPORT LIBRARIES
# ============================================================

import os
import re
import glob
import json
import random
import shutil
import joblib
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import kagglehub

from IPython.display import display

from scipy.sparse import hstack

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)

warnings.filterwarnings("ignore")


# ============================================================
# 3. GLOBAL SETTINGS
# ============================================================

SEED = 42

random.seed(SEED)
np.random.seed(SEED)

SAVE_DIR = "/content/top3_linear_svm_plutchik_8"
os.makedirs(SAVE_DIR, exist_ok=True)


# ============================================================
# 4. DOWNLOAD DATASET
# ============================================================

path = kagglehub.dataset_download(
    "atharvjairath/empathetic-dialogues-facebook-ai"
)

print("Path to dataset files:", path)

csv_files = glob.glob(path + "/**/*.csv", recursive=True)

print("\nCSV files found:")
for file in csv_files:
    print(file)

if len(csv_files) == 0:
    raise FileNotFoundError("No CSV files found.")


# ============================================================
# 5. LOAD DATASET
# ============================================================

selected_file = None

for file in csv_files:
    temp_df = pd.read_csv(file)

    if (
        "emotion" in temp_df.columns
        and "Situation" in temp_df.columns
        and "empathetic_dialogues" in temp_df.columns
    ):
        selected_file = file
        break

if selected_file is None:
    raise ValueError("No valid CSV file found.")

df = pd.read_csv(selected_file)

print("\nSelected CSV:", selected_file)
print("Original shape:", df.shape)
print("Columns:", list(df.columns))

display(df.head())


# ============================================================
# 6. MAP EMPATHETIC DIALOGUES LABELS TO PLUTCHIK 8
# ============================================================

PLUTCHIK_MAP = {
    # JOY
    "joyful": "joy",
    "excited": "joy",
    "content": "joy",
    "grateful": "joy",
    "proud": "joy",

    # TRUST
    "trusting": "trust",
    "faithful": "trust",
    "caring": "trust",
    "confident": "trust",

    # FEAR
    "afraid": "fear",
    "terrified": "fear",
    "anxious": "fear",
    "apprehensive": "fear",

    # SURPRISE
    "surprised": "surprise",
    "impressed": "surprise",

    # SADNESS
    "sad": "sadness",
    "lonely": "sadness",
    "devastated": "sadness",
    "disappointed": "sadness",
    "sentimental": "sadness",
    "nostalgic": "sadness",
    "guilty": "sadness",
    "ashamed": "sadness",
    "embarrassed": "sadness",

    # DISGUST
    "disgusted": "disgust",

    # ANGER
    "angry": "anger",
    "furious": "anger",
    "annoyed": "anger",
    "jealous": "anger",

    # ANTICIPATION
    "anticipating": "anticipation",
    "hopeful": "anticipation",
    "prepared": "anticipation"
}


# ============================================================
# 7. CLEAN DATASET
# ============================================================

df = df.loc[:, ~df.columns.str.contains("^Unnamed")]

df = df.rename(columns={
    "emotion": "original_emotion"
})

df = df.dropna(
    subset=[
        "Situation",
        "empathetic_dialogues",
        "original_emotion"
    ]
)

df["original_emotion"] = (
    df["original_emotion"]
    .astype(str)
    .str.lower()
    .str.strip()
)

before_filter = len(df)

df = df[
    df["original_emotion"].isin(PLUTCHIK_MAP.keys())
].copy()

after_filter = len(df)

print("\nRemoved unmapped labels:", before_filter - after_filter)

df["plutchik_label"] = df["original_emotion"].map(PLUTCHIK_MAP)


# ============================================================
# 8. EXTRACT CUSTOMER TEXT ONLY
# ============================================================

def extract_customer_only(dialogue):
    dialogue = str(dialogue)

    matches = re.findall(
        r"Customer\s*:\s*(.*?)(?=\n?Agent\s*:|$)",
        dialogue,
        flags=re.IGNORECASE | re.DOTALL
    )

    if len(matches) == 0:
        return dialogue

    return " ".join(matches).strip()


df["customer_text"] = df["empathetic_dialogues"].apply(extract_customer_only)


# ============================================================
# 9. CREATE LEAKAGE-SAFE SITUATION GROUP
# ============================================================

df["situation_group"] = (
    df["Situation"]
    .astype(str)
    .str.lower()
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)


# ============================================================
# 10. REMOVE CONFLICTING GROUPS
# ============================================================

group_label_counts = df.groupby("situation_group")["plutchik_label"].nunique()
conflicting_groups = group_label_counts[group_label_counts > 1].index

print("Conflicting situation groups removed:", len(conflicting_groups))

df = df[
    ~df["situation_group"].isin(conflicting_groups)
].copy()


# ============================================================
# 11. FINAL TEXT INPUT
# ============================================================

df["text"] = (
    df["Situation"].astype(str).str.strip()
    + " "
    + df["customer_text"].astype(str).str.strip()
)

df = df[
    [
        "text",
        "original_emotion",
        "plutchik_label",
        "situation_group"
    ]
].copy()

df = df[df["text"].str.len() > 0]

df = df.drop_duplicates(
    subset=[
        "text",
        "original_emotion",
        "plutchik_label",
        "situation_group"
    ]
)

print("\nFinal cleaned shape:", df.shape)
print("Number of Plutchik labels:", df["plutchik_label"].nunique())

print("\nPlutchik label distribution:")
label_distribution = df["plutchik_label"].value_counts().sort_index()
print(label_distribution)

display(df.head())


# ============================================================
# 12. LABEL DISTRIBUTION VISUALIZATION
# ============================================================

plt.figure(figsize=(10, 6))
sns.barplot(
    x=label_distribution.index,
    y=label_distribution.values
)
plt.title("Plutchik Emotion Label Distribution")
plt.xlabel("Emotion Category")
plt.ylabel("Number of Samples")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(f"{SAVE_DIR}/label_distribution.png", dpi=300)
plt.show()


# ============================================================
# 13. GROUP-SAFE TRAIN / VALIDATION / TEST SPLIT
# ============================================================

group_df = (
    df[["situation_group", "plutchik_label"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

train_groups, temp_groups = train_test_split(
    group_df,
    test_size=0.30,
    random_state=SEED,
    stratify=group_df["plutchik_label"]
)

valid_groups, test_groups = train_test_split(
    temp_groups,
    test_size=0.50,
    random_state=SEED,
    stratify=temp_groups["plutchik_label"]
)

train_df = df[df["situation_group"].isin(train_groups["situation_group"])].copy()
valid_df = df[df["situation_group"].isin(valid_groups["situation_group"])].copy()
test_df = df[df["situation_group"].isin(test_groups["situation_group"])].copy()

train_set = set(train_df["situation_group"])
valid_set = set(valid_df["situation_group"])
test_set = set(test_df["situation_group"])

print("\nTrain shape:", train_df.shape)
print("Validation shape:", valid_df.shape)
print("Test shape:", test_df.shape)

print("\nLeakage check:")
print("Train-Validation overlap:", len(train_set & valid_set))
print("Train-Test overlap:", len(train_set & test_set))
print("Validation-Test overlap:", len(valid_set & test_set))

assert len(train_set & valid_set) == 0
assert len(train_set & test_set) == 0
assert len(valid_set & test_set) == 0


# ============================================================
# 14. LABEL ENCODING
# ============================================================

label_encoder = LabelEncoder()
label_encoder.fit(train_df["plutchik_label"])

y_train = label_encoder.transform(train_df["plutchik_label"])
y_valid = label_encoder.transform(valid_df["plutchik_label"])
y_test = label_encoder.transform(test_df["plutchik_label"])

class_names = label_encoder.classes_
num_classes = len(class_names)

id2label = {
    int(i): label
    for i, label in enumerate(class_names)
}

label2id = {
    label: int(i)
    for i, label in enumerate(class_names)
}

print("\nLabels:", class_names)
print("id2label:", id2label)


# ============================================================
# 15. PREPARE TEXT ARRAYS
# ============================================================

X_train_text = train_df["text"].astype(str).values
X_valid_text = valid_df["text"].astype(str).values
X_test_text = test_df["text"].astype(str).values

print("\nTrain samples:", len(X_train_text))
print("Validation samples:", len(X_valid_text))
print("Test samples:", len(X_test_text))


# ============================================================
# 16. TF-IDF FEATURE EXTRACTION
# ============================================================

word_vectorizer = TfidfVectorizer(
    analyzer="word",
    ngram_range=(1, 3),
    max_features=250000,
    min_df=2,
    max_df=0.95,
    sublinear_tf=True,
    lowercase=True,
    strip_accents="unicode"
)

char_vectorizer = TfidfVectorizer(
    analyzer="char_wb",
    ngram_range=(3, 5),
    max_features=100000,
    min_df=2,
    max_df=0.95,
    sublinear_tf=True,
    lowercase=True,
    strip_accents="unicode"
)

X_train_word = word_vectorizer.fit_transform(X_train_text)
X_valid_word = word_vectorizer.transform(X_valid_text)
X_test_word = word_vectorizer.transform(X_test_text)

X_train_char = char_vectorizer.fit_transform(X_train_text)
X_valid_char = char_vectorizer.transform(X_valid_text)
X_test_char = char_vectorizer.transform(X_test_text)

X_train = hstack([X_train_word, X_train_char]).tocsr()
X_valid = hstack([X_valid_word, X_valid_char]).tocsr()
X_test = hstack([X_test_word, X_test_char]).tocsr()

print("\nWord TF-IDF train shape:", X_train_word.shape)
print("Char TF-IDF train shape:", X_train_char.shape)
print("Combined train shape:", X_train.shape)
print("Combined validation shape:", X_valid.shape)
print("Combined test shape:", X_test.shape)


# ============================================================
# 17. TOP 3 LINEAR SVM CONFIGURATIONS ONLY
# ============================================================

top3_svm_configs = [
    {
        "rank": 1,
        "C": 0.3,
        "class_weight": "balanced"
    },
    {
        "rank": 2,
        "C": 0.5,
        "class_weight": "balanced"
    },
    {
        "rank": 3,
        "C": 0.3,
        "class_weight": None
    }
]


# ============================================================
# 18. TRAIN TOP 3 MODELS ON TRAIN SET
#     EVALUATE ON VALIDATION SET
# ============================================================

validation_results = []
trained_models = []

for config in top3_svm_configs:
    print("\n" + "=" * 70)
    print(f"Training Top {config['rank']} Linear SVM")
    print("C:", config["C"])
    print("class_weight:", config["class_weight"])
    print("=" * 70)

    model = LinearSVC(
        C=config["C"],
        class_weight=config["class_weight"],
        random_state=SEED,
        max_iter=12000
    )

    model.fit(X_train, y_train)

    valid_pred = model.predict(X_valid)

    valid_accuracy = accuracy_score(y_valid, valid_pred)
    valid_precision_macro = precision_score(
        y_valid,
        valid_pred,
        average="macro",
        zero_division=0
    )
    valid_recall_macro = recall_score(
        y_valid,
        valid_pred,
        average="macro",
        zero_division=0
    )
    valid_f1_macro = f1_score(
        y_valid,
        valid_pred,
        average="macro",
        zero_division=0
    )
    valid_f1_weighted = f1_score(
        y_valid,
        valid_pred,
        average="weighted",
        zero_division=0
    )

    result = {
        "rank": config["rank"],
        "C": config["C"],
        "class_weight": config["class_weight"],
        "valid_accuracy": valid_accuracy,
        "valid_macro_precision": valid_precision_macro,
        "valid_macro_recall": valid_recall_macro,
        "valid_macro_f1": valid_f1_macro,
        "valid_weighted_f1": valid_f1_weighted
    }

    validation_results.append(result)

    trained_models.append({
        "config": config,
        "model": model,
        "validation_result": result
    })

    print("Validation Accuracy:", valid_accuracy)
    print("Validation Macro Precision:", valid_precision_macro)
    print("Validation Macro Recall:", valid_recall_macro)
    print("Validation Macro F1:", valid_f1_macro)
    print("Validation Weighted F1:", valid_f1_weighted)


validation_results_df = pd.DataFrame(validation_results)

validation_results_df = validation_results_df.sort_values(
    by=["valid_macro_f1", "valid_weighted_f1", "valid_accuracy"],
    ascending=False
).reset_index(drop=True)

print("\nTop 3 Validation Results:")
display(validation_results_df)

validation_results_df.to_csv(
    f"{SAVE_DIR}/top3_validation_results.csv",
    index=False
)


# ============================================================
# 19. VALIDATION RESULT VISUALIZATION
# ============================================================

validation_plot_df = validation_results_df.copy()

validation_plot_df["model_config"] = validation_plot_df.apply(
    lambda row: f"C={row['C']}, weight={row['class_weight']}",
    axis=1
)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=validation_plot_df,
    x="model_config",
    y="valid_macro_f1"
)
plt.title("Top 3 Linear SVM Validation Macro F1")
plt.xlabel("Model Configuration")
plt.ylabel("Validation Macro F1")
plt.ylim(0, 1)
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig(f"{SAVE_DIR}/top3_validation_macro_f1.png", dpi=300)
plt.show()


# ============================================================
# 20. SELECT BEST MODEL BASED ON VALIDATION MACRO F1
# ============================================================

best_result = validation_results_df.iloc[0]

best_model_info = None

for item in trained_models:
    if (
        item["config"]["C"] == best_result["C"]
        and item["config"]["class_weight"] == best_result["class_weight"]
    ):
        best_model_info = item
        break

best_model = best_model_info["model"]

print("\n" + "=" * 70)
print("BEST SELECTED MODEL")
print("=" * 70)
print("C:", best_result["C"])
print("class_weight:", best_result["class_weight"])
print("Validation Accuracy:", best_result["valid_accuracy"])
print("Validation Macro F1:", best_result["valid_macro_f1"])
print("Validation Weighted F1:", best_result["valid_weighted_f1"])


# ============================================================
# 21. TEST EVALUATION
# ============================================================

y_test_pred = best_model.predict(X_test)

test_accuracy = accuracy_score(y_test, y_test_pred)

test_macro_precision = precision_score(
    y_test,
    y_test_pred,
    average="macro",
    zero_division=0
)

test_macro_recall = recall_score(
    y_test,
    y_test_pred,
    average="macro",
    zero_division=0
)

test_macro_f1 = f1_score(
    y_test,
    y_test_pred,
    average="macro",
    zero_division=0
)

test_weighted_precision = precision_score(
    y_test,
    y_test_pred,
    average="weighted",
    zero_division=0
)

test_weighted_recall = recall_score(
    y_test,
    y_test_pred,
    average="weighted",
    zero_division=0
)

test_weighted_f1 = f1_score(
    y_test,
    y_test_pred,
    average="weighted",
    zero_division=0
)

print("\n" + "=" * 70)
print("FINAL TEST RESULTS")
print("=" * 70)
print("Test Accuracy:", test_accuracy)
print("Test Macro Precision:", test_macro_precision)
print("Test Macro Recall:", test_macro_recall)
print("Test Macro F1:", test_macro_f1)
print("Test Weighted Precision:", test_weighted_precision)
print("Test Weighted Recall:", test_weighted_recall)
print("Test Weighted F1:", test_weighted_f1)


# ============================================================
# 22. CLASSIFICATION REPORT
# ============================================================

report_dict = classification_report(
    y_test,
    y_test_pred,
    labels=np.arange(num_classes),
    target_names=class_names,
    digits=4,
    zero_division=0,
    output_dict=True
)

report_df = pd.DataFrame(report_dict).transpose()

print("\nClassification Report:")
display(report_df)

print("\nClassification Report Text:")
print(
    classification_report(
        y_test,
        y_test_pred,
        labels=np.arange(num_classes),
        target_names=class_names,
        digits=4,
        zero_division=0
    )
)

report_df.to_csv(f"{SAVE_DIR}/classification_report.csv")


# ============================================================
# 23. PER-CATEGORY ACCURACY, RECALL, F1, SUPPORT
# ============================================================

cm = confusion_matrix(
    y_test,
    y_test_pred,
    labels=np.arange(num_classes)
)

per_class_support = cm.sum(axis=1)
per_class_correct = np.diag(cm)

per_class_accuracy = np.divide(
    per_class_correct,
    per_class_support,
    out=np.zeros_like(per_class_correct, dtype=float),
    where=per_class_support != 0
)

per_category_df = report_df.loc[class_names][
    [
        "precision",
        "recall",
        "f1-score",
        "support"
    ]
].copy()

per_category_df.insert(
    0,
    "class_accuracy",
    per_class_accuracy
)

per_category_df = per_category_df.rename(
    columns={
        "precision": "precision",
        "recall": "recall",
        "f1-score": "f1_score",
        "support": "support"
    }
)

print("\nPer-Category Evaluation Results:")
display(per_category_df)

per_category_df.to_csv(
    f"{SAVE_DIR}/per_category_results.csv"
)


# ============================================================
# 24. OVERALL METRICS SUMMARY
# ============================================================

overall_metrics = {
    "accuracy": test_accuracy,
    "macro_precision": test_macro_precision,
    "macro_recall": test_macro_recall,
    "macro_f1": test_macro_f1,
    "weighted_precision": test_weighted_precision,
    "weighted_recall": test_weighted_recall,
    "weighted_f1": test_weighted_f1
}

overall_metrics_df = pd.DataFrame(
    list(overall_metrics.items()),
    columns=["Metric", "Score"]
)

print("\nOverall Test Metrics:")
display(overall_metrics_df)

overall_metrics_df.to_csv(
    f"{SAVE_DIR}/overall_test_metrics.csv",
    index=False
)

with open(f"{SAVE_DIR}/overall_test_metrics.json", "w") as f:
    json.dump(overall_metrics, f, indent=4)


# ============================================================
# 25. OVERALL METRICS VISUALIZATION
# ============================================================

plt.figure(figsize=(10, 6))
sns.barplot(
    data=overall_metrics_df,
    x="Metric",
    y="Score"
)
plt.title("Overall Test Metrics")
plt.xlabel("Metric")
plt.ylabel("Score")
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(f"{SAVE_DIR}/overall_test_metrics.png", dpi=300)
plt.show()


# ============================================================
# 26. PER-CATEGORY F1 VISUALIZATION
# ============================================================

plt.figure(figsize=(10, 6))
sns.barplot(
    x=per_category_df.index,
    y=per_category_df["f1_score"]
)
plt.title("Per-Category F1 Score")
plt.xlabel("Emotion Category")
plt.ylabel("F1 Score")
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(f"{SAVE_DIR}/per_category_f1_score.png", dpi=300)
plt.show()


# ============================================================
# 27. SEABORN RAW CONFUSION MATRIX
# ============================================================

cm_df = pd.DataFrame(
    cm,
    index=class_names,
    columns=class_names
)

plt.figure(figsize=(10, 8))
sns.heatmap(
    cm_df,
    annot=True,
    fmt="d",
    cmap="Blues"
)
plt.title("Raw Confusion Matrix - Linear SVM")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.savefig(f"{SAVE_DIR}/raw_confusion_matrix_seaborn.png", dpi=300)
plt.show()

cm_df.to_csv(f"{SAVE_DIR}/raw_confusion_matrix.csv")


# ============================================================
# 28. SEABORN NORMALIZED CONFUSION MATRIX
# ============================================================

cm_normalized = np.divide(
    cm.astype(float),
    cm.sum(axis=1, keepdims=True),
    out=np.zeros_like(cm, dtype=float),
    where=cm.sum(axis=1, keepdims=True) != 0
)

cm_normalized_df = pd.DataFrame(
    cm_normalized,
    index=class_names,
    columns=class_names
)

plt.figure(figsize=(10, 8))
sns.heatmap(
    cm_normalized_df,
    annot=True,
    fmt=".2f",
    cmap="Blues"
)
plt.title("Normalized Confusion Matrix - Linear SVM")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.savefig(f"{SAVE_DIR}/normalized_confusion_matrix_seaborn.png", dpi=300)
plt.show()

cm_normalized_df.to_csv(f"{SAVE_DIR}/normalized_confusion_matrix.csv")


# ============================================================
# 29. SAVE MODEL AND ARTIFACTS
# ============================================================

joblib.dump(best_model, f"{SAVE_DIR}/linear_svm_model.pkl")
joblib.dump(word_vectorizer, f"{SAVE_DIR}/word_tfidf_vectorizer.pkl")
joblib.dump(char_vectorizer, f"{SAVE_DIR}/char_tfidf_vectorizer.pkl")
joblib.dump(label_encoder, f"{SAVE_DIR}/label_encoder.pkl")

with open(f"{SAVE_DIR}/id2label.json", "w") as f:
    json.dump(id2label, f, indent=4)

with open(f"{SAVE_DIR}/label2id.json", "w") as f:
    json.dump(label2id, f, indent=4)

best_model_config = {
    "C": float(best_result["C"]),
    "class_weight": best_result["class_weight"],
    "validation_accuracy": float(best_result["valid_accuracy"]),
    "validation_macro_f1": float(best_result["valid_macro_f1"]),
    "validation_weighted_f1": float(best_result["valid_weighted_f1"])
}

with open(f"{SAVE_DIR}/best_model_config.json", "w") as f:
    json.dump(best_model_config, f, indent=4)


# ============================================================
# 30. SAMPLE PREDICTION FUNCTION
# ============================================================

def predict_emotion_svm(text, top_k=5):
    X_word = word_vectorizer.transform([text])
    X_char = char_vectorizer.transform([text])
    X_input = hstack([X_word, X_char]).tocsr()

    scores = best_model.decision_function(X_input)[0]

    top_indices = np.argsort(scores)[-top_k:][::-1]

    results = []

    for idx in top_indices:
        results.append({
            "emotion": label_encoder.inverse_transform([idx])[0],
            "svm_decision_score": float(scores[idx])
        })

    return results


sample_text = "I feel alone, guilty, nervous, and disappointed with myself."

sample_results = predict_emotion_svm(
    sample_text,
    top_k=5
)

print("\nSample Input:")
print(sample_text)

print("\nTop SVM Emotion Scores:")
for item in sample_results:
    print(f"{item['emotion']}: {item['svm_decision_score']:.4f}")


# ============================================================
# 31. ZIP ALL OUTPUTS FOR DOWNLOAD
# ============================================================

zip_path = shutil.make_archive(
    base_name=SAVE_DIR,
    format="zip",
    root_dir="/content",
    base_dir="top3_linear_svm_plutchik_8"
)

print("\nSaved files to:")
print(SAVE_DIR)

print("\nDownload file:")
print(zip_path)

try:
    from google.colab import files
    files.download(zip_path)
except Exception:
    print("\nManual download path:")
    print(zip_path)
```
